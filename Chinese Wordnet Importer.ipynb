Imports and configuration.
==========================

We need 
- the path to the SQLite3 file, 
- the padagraph URL 
- and the Key to use the bot API (get one at PDG_HOST/account/me/generate_auth_token )

```python
import sqlite3
from botapi import Botagraph, BotApiError
from reliure.types import Text

GRAPHNAME = "Chinese Wordnet"
DBPATH    = "./cwn_dirty.sqlite"
PDG_HOST  = "http://lope.padagraph.io"

```

Definitions of the datatypes and extraction from the SQLite DB
====================

```python
from collections import namedtuple

Sense = namedtuple("Sense", "id lemma_id definition synset")
Lemma = namedtuple("Lemma", "id pinyin zhuyin wordform")
Synset = namedtuple("Synset", "id gloss pwn_gloss pwn_word")
Relation = namedtuple("Relation", "source target reltype")
```

```python
db = sqlite3.connect(DBPATH)

cursor = db.execute("SELECT sense_id, lemma_id, sense_def, synset_id from cwn_sense")
senses = {}
for row in cursor.fetchall():
    s = Sense(*row)
    senses[s.id] = s

cursor = db.execute("SELECT lemma_id, cwn_pinyin, cwn_zhuyin, lemma_type from cwn_lemma")
lemmas = {}
for row in cursor.fetchall():
    l = Lemma(*row)
    lemmas[l.id] = l

cursor = db.execute("SELECT id, gloss, pwn_gloss, pwn_word from cwn_goodSynset")
synsets = {}
for row in cursor.fetchall():
    ss = Synset(*row)
    synsets[ss.id] =  ss

    
cursor = db.execute("SELECT cwn_id, rel_cwnid, rel_type from cwn_relation")
relations = [Relation(*x) for x in cursor.fetchall()]
```

Definitions of NodeTypes and EdgeTypes to be used in Padagraph
==============================================================


```python
NodeType = namedtuple("NodeType", "name description properties")
EdgeType = namedtuple("EdgeType", "name description properties")

SenseNode = NodeType("Sense",
                     "some atomic sense",
                     {"label": Text(), 
                      "shape": Text(),
                      "id": Text(),
                      "definition": Text()})

LemmaNode = NodeType("Lemma",
                     "",
                     {"label": Text(),
                      "id": Text(),
                      "pinyin": Text(),
                      "zhuyin": Text(),
                      "wordform": Text()})

SynsetNode = NodeType("Synset",
                     "mapping to PWN",
                     {"label": Text(), 
                      "shape": Text(),
                      "id": Text(),
                      "gloss": Text(),
                      "PWN": Text(),
                      "word": Text()})

SynonymRel = EdgeType("synonym","", {})
AntonymRel = EdgeType("antonym","", {})
HyponymRel = EdgeType("hyponym","", {})
HyperonymRel = EdgeType("hyperonym","", {})

# TODO: what do those symbols stand for
sharpRel = EdgeType("#","", {})
percentRel = EdgeType("%","", {})
andRel = EdgeType("&","", {})
upRel = EdgeType("^","", {})





hasSenseRel = EdgeType("has sense", "", {})
containsRel = EdgeType("contains", "", {})
```

Creating the graph on the server And adding Types
=================================================





```python
bot = Botagraph(PDG_HOST, PDG_KEY)
bot.create_graph(GRAPHNAME, {'description': "", 'image':"http://lope.linguistics.ntu.edu.tw/images/logo.jpg", 'tags':['chinese', 'wordnet']})

for t in [SenseNode, LemmaNode, SynsetNode]:
    bot.post_nodetype(GRAPHNAME, t.name, t.description, t.properties)

for t in [SynonymRel, AntonymRel, HyponymRel, HyperonymRel, hasSenseRel, containsRel]:
    bot.post_edgetype(GRAPHNAME, t.name, t.description, t.properties)


```

```python

schema = bot.get_schema(GRAPHNAME)['schema']
nodetypes = { n['name']:n for n in schema['nodetypes'] }
edgetypes = { e['name']:e for e in schema['edgetypes'] }
```

Creating the Nodes
==================

```python

idx = {}
# Lemmas

none_as_str = lambda x: x if x is not None else ""

def build_lemma_properties(lemma):
    return {'nodetype': nodetypes['Lemma']['uuid'],
            'properties':  {"label": lemma.wordform,
                            "shape": 'circle',
                            "id": str(lemma.id),
                            "pinyin": none_as_str(lemma.pinyin) ,
                            "zhuyin": none_as_str(lemma.zhuyin),
                            "wordform": lemma.wordform} }

for node, uuid in bot.post_nodes(GRAPHNAME, iter(build_lemma_properties(x) for x in lemmas.values())):
    idx[('Lemma',node['properties']['id'])] = uuid


# Senses
        
def build_sense_properties(sense):
    label = sense.definition[:4]
    if len(sense.definition) > 4:
        label += u"…"
    return {'nodetype': nodetypes['Sense']['uuid'],
            'properties': {"label": label, 
                      "shape": 'triangle',
                      "id": str(sense.id),
                      "definition": sense.definition} }

for node, uuid in bot.post_nodes(GRAPHNAME, iter(build_sense_properties(x) for x in senses.values())):
    idx[('Sense',node['properties']['id'])] = uuid

  
    
# Synsets
    
def build_synset_properties(synset):
    if synset.pwn_word:
        label = synset.pwn_word
    else:
        label = none_as_str(synset.gloss)[:4]
        if len(none_as_str(synset.gloss)) > 4:
            label += u"…"
    return {'nodetype': nodetypes['Synset']['uuid'],
            'properties': {"label": label,
                      "shape": "square",
                      "id": str(synset.id),
                      "gloss": none_as_str(synset.gloss),
                      "PWN": none_as_str(synset.pwn_gloss),
                      "word": none_as_str(synset.pwn_word)}}

for node, uuid in bot.post_nodes(GRAPHNAME, iter(build_synset_properties(x) for x in synsets.values())):
    idx[('Synset',node['properties']['id'])] = uuid
```

Creating the Edges
==================

```python
# lemma has sense



rels = []
for s in senses.values():
    try:
        sense_uuid = idx[('Sense',s.id)]
        lemma_uuid = idx[('Lemma',s.lemma_id)]
        rels.append({'edgetype':edgetypes['has sense']['uuid'],
                 'source': lemma_uuid,
                 'target': sense_uuid,
                 'properties':{}})
        if s.synset is not None:
            synset_uuid = idx[('Synset',s.synset)]
            rels.append({'edgetype':edgetypes['contains']['uuid'],
                 'source': synset_uuid,
                 'target': sense_uuid,
                 'properties':{}})
    except:
        pass
        # print "error",s.id, s.lemma_id, s.synset
[ x for x in bot.post_edges(GRAPHNAME, iter(rels))]

# wordnet relations

def relation_mapping(r):
    if r == '=':
        return 'synonym'
    elif r == '!':
        return 'antonym'
    elif r == '~':
        return 'hyponym'
    elif r == '@':
        return 'hyperonym'
    else:
        return r

rels = []
for r in relations:
    if r.source and r.target:
        try:
            src_uuid = idx[('Sense',r.source[:8])]
            tgt_uuid = idx[('Sense',r.target)]
            rels.append({'edgetype': edgetypes[relation_mapping(r.reltype)]['uuid'],
                'source': src_uuid,
                'target': tgt_uuid,
                'properties': {}})
        except:
            print "failed",r
   
[x for x in bot.post_edges(GRAPHNAME, iter(rels))]
```
